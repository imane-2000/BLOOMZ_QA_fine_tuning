```python
from datasets import load_dataset
```

```python
CSqa=load_dataset("commonsense_qa")
```

```python
import pandas as pd

common=pd.DataFrame(CSqa['validation'])
```

```python
common
```

```python
def format_choices(choices):
    result = "\n".join(f"{label} {text}" for label, text in zip(choices['label'], choices['text']))
    return result
common['formatted_choices'] = common['choices'].apply(format_choices)
```

```python
common
```

```python
def buildprompt(data):
    prompt = {}
    prompt['text'] = "answer the following question by specifying the answer key: \n"+ data['question'] +"\n" + data['formatted_choices']+ "\nAnswer: " + data['answerKey']
    return prompt

common['prompt'] = common.apply(buildprompt, axis=1)
```

```python
common['prompt']
```

```python
import json
result = common['prompt'].to_list()
with open('prompts.json', 'w') as outfile:
    json.dump(result, outfile, ensure_ascii=False)
```

```python
dataset = load_dataset("json", data_files="prompts.json")
```

```python
from transformers import BloomTokenizerFast, BloomForCausalLM, TrainingArguments, Trainer
import torch

checkpoint = "bigscience/bloomz-560m"

tokenizer = BloomTokenizerFast.from_pretrained(checkpoint)
model = BloomForCausalLM.from_pretrained(checkpoint).to("cuda")
```

```python
def prepare_train_data(data):
    # prompt + completion
    text_input = data['text']
    # tokenize the input (prompt + completion) text
    tokenized_input = tokenizer(
    text_input,
    return_tensors='pt',
    padding='max_length',  # Pad to the maximum length
    max_length=93  # Set the maximum length
)
    # generative models: labels are the same as the input
    tokenized_input['labels'] = tokenized_input['input_ids']
    return tokenized_input

train_dataset = dataset['train'].map(prepare_train_data,
                                     batched=True,
                                     remove_columns=["text"])
```

```python
training_arguments = TrainingArguments(
    'FT-bloom-560m',
    learning_rate=2e-5,
    per_device_train_batch_size=2,
    num_train_epochs=4,
    weight_decay=0.01,
    fp16=True,
    optim="adafactor",
    gradient_accumulation_steps=4,
    gradient_checkpointing=True
)
```

```python
trainer = Trainer(
    model = model,
    args = training_arguments,
    train_dataset = train_dataset
)
```

```python
trainer.train()
```

```python
trainer.save_model()
```

```python
results = []
for i, P in enumerate(commontest["prompt"]):
    prompt = P
    inputs = tokenizer(prompt, return_tensors="pt")
    inputs = {k: v.to('cuda:0') for k, v in inputs.items()}  # Move inputs to cuda:0

    # Ensure the model is on the same device (cuda:0)
    model.to('cuda:0')

    result = (tokenizer.decode(model.generate(inputs["input_ids"],
                       max_length=len(prompt) + 1,
                       do_sample=True,
                       top_k=50,
                       top_p=0.9
                      )[0]))
    results.append(result[len(prompt)+1])
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
def generate_evaluation_report(df, label_column, prediction_column, labels):
    # Calculate accuracy
    accuracy = accuracy_score(df[label_column], df[prediction_column])

    # Calculate confusion matrix
    conf_matrix = confusion_matrix(df[label_column], df[prediction_column], labels=labels)
    confusion_df = pd.DataFrame(conf_matrix, columns=labels, index=labels)

    # Generate a classification report
    class_report = classification_report(df[label_column], df[prediction_column], labels=labels, output_dict=True)

    evaluation_report = {
        'Accuracy': accuracy,
        'Confusion Matrix': confusion_df,
        'Classification Report': class_report
    }

    return evaluation_report

```

```python
def print_evaluation_report(report):
    print(f'Accuracy: {report["Accuracy"]:.2f}\n')
    print("Confusion Matrix:")
    print(report['Confusion Matrix'])
    print("\nClassification Report:")
    for label, metrics in report['Classification Report'].items():
        if label not in ['accuracy', 'macro avg', 'weighted avg']:
            print(f"Label {label} - Precision: {metrics['precision']:.2f}, Recall: {metrics['recall']:.2f}, F1-Score: {metrics['f1-score']:.2f}")
```

```python
labels = ['A', 'B', 'C', 'D', 'E', 'M']
```

```python
report_ft = generate_evaluation_report(commontest, 'answerKey', 'BloomZ-ft', labels)
report_560m=generate_evaluation_report(commontest, 'answerKey', 'BloomZ-560m', labels)
report_3b=generate_evaluation_report(commontest, 'answerKey', 'BloomZ-3b', labels)
```

```python
print("Evaluation Report for the fine-tuned model:")
print_evaluation_report(report_ft)
```

```python
print("Evaluation Report for BLOOMZ-560m model:")
print_evaluation_report(report_560m)
```

```python
print("Evaluation Report for BLOOMZ-3b model:")
print_evaluation_report(report_3b)
```
